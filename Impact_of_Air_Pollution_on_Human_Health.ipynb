# **Data Cleaning and Preparation**

Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly .express as px
from IPython.core.display import HTML 

%matplotlib inline
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (17,5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

Load Dataset

```python
df1 = pd.read_csv("absolute-number-of-deaths-from-ambient-particulate-air-pollution.csv")
df2 = pd.read_csv('death-rate-from-air-pollution-per-100000.csv')
df3 = pd.read_csv('death-rates-from-air-pollution.csv')
df4 = pd.read_csv('share-deaths-indoor-pollution.csv')
df5 = pd.read_csv('outdoor-pollution-death-rate.csv')
```

```python
df1
```

Rename requred column

```python
df_1 = df1.rename(columns = {'Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: All Ages (Number)':'Total Deaths for outdoor Air Polution','Deaths - Cause: All causes - Risk: Household air pollution from solid fuels - Sex: Both - Age: All Ages (Number)':'Total Deaths for Household Air Polution from solid fuels','Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: All Ages (Number)':'Total Deaths from Air Polution'})
```

```python
df_1
```

```python
new_df_1 = df_1[['Entity','Year','Total Deaths for outdoor Air Polution','Total Deaths for Household Air Polution from solid fuels','Total Deaths from Air Polution']]
```

```python
df_2 = df2.rename(columns={'Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Rate)':'Deaths rate from Air Polution per 100000'})
df_3 = df3.rename(columns={'Deaths - Cause: All causes - Risk: Ambient particulate matter pollution - Sex: Both - Age: Age-standardized (Rate)':'Deaths rate from Ambient particulate matter polution','Deaths - Cause: All causes - Risk: Ambient ozone pollution - Sex: Both - Age: Age-standardized (Rate)':'Deaths rate from Ambient ozone pollution',}).drop(['Deaths - Cause: All causes - Risk: Household air pollution from solid fuels - Sex: Both - Age: Age-standardized (Rate)','Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Rate)','Code'], axis=1)
df_4 = df4.rename(columns={'Deaths - Cause: All causes - Risk: Household air pollution from solid fuels - Sex: Both - Age: Age-standardized (Percent)':'Deaths for Household air pollution from solid fuels (Percent)'})
df_5 = df5.rename(columns={'Deaths - Cause: All causes - Risk: Outdoor air pollution - OWID - Sex: Both - Age: Age-standardized (Rate)':'Deaths for Outdoor air polution (per 100K)'})
```
